```
zoom_x = 2.0  # horizontal zoom
zoom_y = 2.0  # vertical zoom
mat = fitz.Matrix(zoom_x, zoom_y)  # zoom factor 2 in each dimension

for name in glob.glob('PDF_letters/*.pdf'):
    doc = fitz.open(name)  # open document
    for page in doc:  # iterate through the pages
        pix = page.get_pixmap(matrix=mat)  # render page to an image
        pix.save(name.split(r'.')[0] + '.png')
    #images.save(name.split(r'/')[1] + '.png', 'PNG')
```

```
for name in glob.glob('PDF_letters/*.png'):
# Import the image
    file_name = name
    
    img = Image.open(file_name) # get image
    pixels = img.load() # create the pixel map

    for i in range(img.size[0]): # for every pixel:
        for j in range(img.size[1]):
            if pixels[i,j] != (0,0,0): # if not black:
                pixels[i,j] = (255, 255, 255) # change to white

    width, height = img.size
    print(width, height)

    # Setting the points for cropped image
    left = 130
    top = 195
    right = 1090
    bottom = 1155
    
    # Cropped image of above dimension
    # (It will not change original image)
    im1 = img.crop((left, top, right, bottom))
    
    # Shows the image in image viewer
    print(im1.size)
    #resize_image = im1.resize((10,10))
    im1.show()
    numpydata = np.asarray(im1,  dtype="int32")
    print(numpydata)
    df_list = []
    for index, pixel in enumerate(numpydata):
        temp_list = []
        for color in pixel:
            temp_list.append(color[0])
        df_list.append(temp_list)
    #print(df_list)
    df = pd.DataFrame(df_list)
    df = df.replace([255, 254], 0)
    df = df.replace(num_list, 1)
    #print(df)
```
