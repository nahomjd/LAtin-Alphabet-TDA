```
dgmLR = [None]*26
for i in range(26):    
    letter_one_line=letters[i,:]
    # initialize matrix of size 10x10 with all values 100    
    letter=np.full((10, 10), 100)
    # convert one line letter to 10x10 matrix replacing zeros with 100
    for k in range(1,101):
        if letter_one_line[k]==1.0:            
            row=int((k-1)/10)             
            column=(k-1)%10            
            letter[row,column]=k%10    
    dgmLR[i] = lower_star_img(letter)
```

```
#diagnal scanning
#diagonal scanning-from-upper-left
#Column+row-1
dgmSUL = [None]*26
for i in range(26):    
    letter_one_line=letters[i,:]
    # initialize matrix of size 10x10 with all values 100    
    letter=np.full((10, 10), 100)
    # convert one line letter to 10x10 matrix replacing zeros with 100
    for k in range(1,101):
        if letter_one_line[k]==1.0:            
            row=int((k-1)/10)             
            column=(k-1)%10            
            letter[row,column]=9-k%10 + 9-int((k-1)/10)+1
    print(letter, '\n')         
    dgmSUL[i] = lower_star_img(letter)
```

```
dgmCEN = [None]*26
for i in range(26):    
    letter_one_line=letters[i,:]
    # initialize matrix of size 10x10 with all values 100    
    letter=np.full((10, 10), 100)
    # convert one line letter to 10x10 matrix replacing zeros with 100
    for k in range(1,101):
        if letter_one_line[k]==1.0:            
            row=int((k-1)/10)             
            column=(k-1)%10            
            letter[row,column]=max(max(9-row-5+1,9-column-5+1),max(row-5+1,column-5+1))
    print(letter, '\n')
    dgmSUL[i] = lower_star_img(letter)
```
